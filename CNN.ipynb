```python
from keras.models import Sequential
from keras.layers import Convolution2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense
import matplotlib.pyplot as plt
from PIL import Image
from scipy import ndimage
import matplotlib.image as mpimg
import cv2
import os
import glob
import pandas as pd
import numpy as np
```

# Creating dataset of available images

```python
img_dir = "images" # Enter Directory of all images 
data_path = os.path.join(img_dir,'*g')
files = glob.glob(data_path)
images=[]

for i in range(1646):

    img = cv2.imread(files[1])
    print(img)
    images.append(img)
    
```

```python
images = np.array(images)
```

```python
images.shape
```

```python
labels_train = pd.read_csv('train.csv')
labels_test =  pd.read_csv('test.csv')
y =labels_train['emergency_or_not']
```

```python
y.shape
```

```python
X=images
```

# defining CNN 

```python
classifier = Sequential()
```

```python
classifier.add(Convolution2D(16,kernel_size = (3,3),input_shape=(224,224,3),activation='relu',strides=(1,1)))
```

```python
classifier.add(MaxPooling2D(pool_size=(2, 2)))
```

```python
classifier.add(Convolution2D(32, kernel_size = (3, 3), activation='relu',strides=(1,1)))
```

```python
classifier.add(MaxPooling2D(pool_size=(2, 2)))
```

```python
classifier.add(Convolution2D(32, kernel_size = (3, 3), activation='relu',strides=(1,1)))
```

```python
classifier.add(MaxPooling2D(pool_size=(2, 2)))
```

```python
# flattening layer
```

```python
classifier.add(Flatten())
```

```python
# dense layers
```

```python
classifier.add(Dense(128, activation='relu'))
```

```python
classifier.add(Dense(64,activation='relu'))
```

```python
classifier.add(Dense(32,activation='relu'))
```

```python
classifier.add(Dense(1,activation='sigmoid'))
```

```python
classifier.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
from keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(monitor='loss', patience=10)
```

# fitting data

```python
classifier.summary()
```

```python
classifier.fit(X,y,epochs=50,batch_size=500,validation_split=0.2,callbacks=[early_stopping])
```
